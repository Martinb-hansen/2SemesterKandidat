```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math as math
data1 = pd.read_csv('data1.csv', sep= ',', header=None)
data1.columns = ["Sample", "Temperature [°C]",  "accX", "accY", "accZ", "rotX", "rotY", "rotZ"]
```

```python
print(data1)
```

```python
data1.head()
accx = data1['accX']
accz = data1['accZ']
roty = data1['rotY']
YnMinusX = 0.0
YnMinusZ = 0.0
alpha = 0.02
g = 9.82
```

```python
plt.plot(accx)
plt.plot(accz)
plt.plot(roty)
plt.legend(["accX", "accZ", "rotY"])
plt.show()
```

```python
ax = math.cos(accx)*g
az = math.sin(accz)*g
YnX = (1-alpha) * YnMinusX + (alpha * ax)
YnMinusX = YnX
YnZ = (1-alpha) * YnMinusZ + (alpha * az)
YnMinusZ = YnZ

angle =  math.atan2(YnX, YnZ)
print(angle)
```
